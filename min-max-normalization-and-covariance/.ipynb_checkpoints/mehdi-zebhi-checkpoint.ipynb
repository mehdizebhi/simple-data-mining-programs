```python
import pandas as pd
import numpy as np

df = pd.read_csv("arrhythmia.csv", header=None)
data = df.to_numpy()

def is_contain_nan(vector):
    return np.isnan(vector).any()

def get_list_of_col_nan(data):
    col = np.size(data, 1)
    list = []
    for i in range(col):
        if is_contain_nan(data[:,i]):
            list.append(i)
    
    return list


def mean(vector):
    sum = 0
    n = 0
    for i in vector:
        if not np.isnan(i):
            sum += i
            n += 1
    
    return sum / n


def set_mean_instead_nan_in_vector(vector):
    m = mean(vector)
    for i in range(len(vector)):
        if np.isnan(vector[i]):
            vector[i] = m
    
    return vector


# a : data without nan (by mean)
def set_mean_instead_nan(data):
    list_of_nan_vector = get_list_of_col_nan(data)
    
    for i in list_of_nan_vector:
        new_vector = set_mean_instead_nan_in_vector(data[:,i])
        data[:,i] = new_vector
        
    return data


def normal(x, min, max, new_min, new_max):
    if max - min == 0:
        return x
    return (((x - min) / (max - min))*(new_max - new_min)) + new_min

# b
def min_max_normalazation(data, new_min, new_max):
    norm = np.vectorize(normal)
    new_data = []
    for col in data.T:
        min_a = min(col)
        max_a = max(col)
        new_data.append(norm(col, min_a, max_a, new_min, new_max))
    
    return np.transpose(np.array(new_data))


# c
def cov(data):
    col = np.size(data, 1)
    new_data = np.zeros((col, col))
    
    for i, a in enumerate(data.T):
        E_A = np.mean(a)
        for j, b in enumerate(data.T):
            E_B = np.mean(b)
            new_data[i,j] = np.mean(a * b) - (E_A * E_B)
    
    return new_data
            



new_data = set_mean_instead_nan(data)
normal_data = min_max_normalazation(new_data, 0, 1)
cov_matrix = cov(normal_data)


print("orginal data:\n")
print(data)
print("\n\n\nnew data (No NaN data, instead mean):\n")
print(new_data)
print("\n\n\nnormal data:\n")
print(normal_data)
print("\n\n\ncov matrix:\n")
print(cov_matrix)


data_frame = pd.DataFrame(cov_matrix)
data_frame.to_excel("cov.xlsx", "Sheet1")

# print("\n\n\n\n")
# print(np.isnan(cov_matrix).any())

# print("\n\n\ncov matrix:\n")
# print(np.cov(normal_data.T))






```
